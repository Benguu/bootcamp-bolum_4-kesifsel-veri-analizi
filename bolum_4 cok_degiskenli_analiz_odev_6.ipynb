```python
#soru_1
#Sınavlardaki performans cinsiyete, kökene ve anne-baba eğitim durumuna göre değişiklik göstermekte midir?

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
from scipy.stats.mstats import winsorize
from sqlalchemy import create_engine
import warnings

warnings.filterwarnings('ignore')
sns.set(style="whitegrid")
pd.options.display.float_format = '{:.3f}'.format
plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize'] = (8,5.5)
```

```python
stperform = pd.read_csv('C:\\Users\\PC\\Desktop\\StudentsPerformance.csv')
stperform.columns = ['cinsiyet', 'etnik_koken', 'aile_egt_seviye', 'ogle_yemegi', 
                              'hazirlik_kursu', 'matematik_notu', 'okuma_notu', 'yazma_notu']
stperform.head()
```

```python
#öncelikle verilerde eksik değer var mı bakalım.

stperform.describe()
stperform.isnull().sum()*100/stperform.shape[0]

#eksik değer yok.
```

```python
#sürekli değişkenlerde aykırı değer var mı bakalım
plt.subplot(1,3,1)
plt.boxplot(stperform['matematik_notu'])
plt.title('Matematik notu')

plt.subplot(1,3,2)
plt.boxplot(stperform['okuma_notu'])
plt.title('okuma notu')

plt.subplot(1,3,3)
plt.boxplot(stperform['yazma_notu'])
plt.title('yazma notu')

plt.show()     

#Her üç değişkende de aykırı değer vardır.
```

```python
# Puan değişkenlerini winsorize yöntemi ile aykırı değerlerini ortadan kaldıralım. 

winsorize_mat = winsorize(stperform['matematik_notu'], (0.10, 0))
winsorize_okuma= winsorize(stperform['okuma_notu'], (0.10, 0))
winsorize_yazma= winsorize(stperform['yazma_notu'], (0.10, 0))
plt.subplot(1,3,1)
plt.boxplot(winsorize_mat)
plt.title('matematik notu')

plt.subplot(1,3,2)
plt.boxplot(winsorize_okuma)
plt.title('okuma notu')

plt.subplot(1,3,3)
plt.boxplot(winsorize_yazma)
plt.title('yazma notu')

plt.show()   
```

```python
cinsiyet = stperform["cinsiyet"].unique()
grup_stperform = stperform.groupby("cinsiyet")
```

```python
pd.options.display.float_format = '{:.15f}'.format
for var in ["matematik_notu", "okuma_notu", "yazma_notu"]:
    karsilastirma = pd.DataFrame(columns=['grup_1', 'grup_2','istatistik', 'p_degeri'])
    print("{} için karşılaştırma".format(var),end='')
    for i in range(0, len(cinsiyet)):
        for j in range(i+1, len(cinsiyet)):
            ttest = stats.ttest_ind(stperform[stperform["cinsiyet"]==cinsiyet[i]][var], 
                                stperform[stperform["cinsiyet"]==cinsiyet[j]][var])
            grup_1 = cinsiyet[i]
            grup_2 = cinsiyet[j]
            istatistik = ttest[0]
            p_degeri = ttest[1]
            
            karsilastirma = karsilastirma.append({"grup_1" : grup_1 ,
                                                  "grup_2" : grup_2 ,
                                                  "istatistik": istatistik , 
                                                  "p_degeri" : p_degeri}, ignore_index=True)
    display(karsilastirma)
```
